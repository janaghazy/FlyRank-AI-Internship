Why?
I want to try something different that actually matters for business - finding
pages that get good visibility but don't get clicks.

The goal: find pages that are underperforming on clicks and be able to spot exactly what needs to be fixed.

**What am I solving?** Finding pages that get high impressions but low clicks,
relative to where they rank.

**What does someone do with this?** SEO team rewrites titles or meta descriptions
on the flagged pages.

**What happens if I'm wrong?**
- Flag a page that's actually fine → waste someone's time
- Miss a page with real potential → lose traffic and revenue

```python
import pandas as pd
df = pd.read_csv("content_refresh_anonymized.csv")

# Show CTR drops by position
pos_ctr = df.groupby("position_tier")["ctr"].mean()
print("CTR by position:")
print(pos_ctr.round(4))

# Show how many visible pages have terrible CTR
visible = df[df["impressions_90d"] >= 500]
low = visible[visible["ctr"] < 0.05]
print(f"\nPages with >500 impressions: {len(visible)}")
print(f"CTR under 5%: {len(low)} ({len(low)/len(visible)*100:.1f}%)")
```

What I can actually say from this data:
- CTR varies significantly by position. Pages in better positions get more clicks.
- About 20% of visible pages get fewer clicks than expected for their position.

What I can't say:
- "Better titles will definitely fix these pages" - I haven't tested that.
- "These findings apply everywhere" - this is one dataset from one set of clients.

   Yes - it tells content teams which pages to look at first.

   Yes - track if CTR improves after changes.
